```python
cols = ['RAM (GB)', 'Storage (GB)']  # columns you want to clean

for col in cols:
    pho[col] = pho[col].replace('NIL', np.nan)                     # Replace NIL with NaN
    pho[col] = pd.to_numeric(pho[col], errors='coerce')            # Convert to numeric
    pho[col] = pho[col].fillna(pho[col].median()) 
```

```python
pho = pho[pho['Camera'].notna()]  # keeps only rows where Camera has a value
```

```python
pho.isnull().sum()
```

```python
pho1 = pho
```

```python
pho1.describe()
```

```python
pho1
```

```python
pho2 = pho1
```

```python
price_cols = ['Actual price', 'Discount price']

for col in price_cols:
    # Remove commas and convert to numeric
     pho1[col] = pho1[col].astype(str)
     pho1[col] = pho1[col].str.replace(',', '', regex=True)
     pho1[col] = pd.to_numeric(pho1[col], errors='coerce')
```

```python
cols = ['Actual price', 'Discount price','Stars','RAM (GB)', 'Storage (GB)']

# Compute correlation
corr = pho1[cols].corr()
sns.heatmap(corr, annot = True )
plt.title('heatmap')
plt.show()
```

```python
brand_counts = pho1['Brand'].value_counts()
print("Brand Frequency Analysis:")
print(brand_counts)
```

```python
# Plotting Brand Frequency
plt.figure(figsize=(12, 6))
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette='viridis')
plt.title('Brand Frequency in Dataset')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.xticks(rotation=90)
plt.show()
```

```python
average_price_per_brand = pho1.groupby('Brand')['Actual price'].mean().sort_values(ascending=False)
print("\nAverage Prices by Brand:")
print(average_price_per_brand)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_per_brand.index, y=average_price_per_brand.values, palette='magma')
plt.title('Average Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

```python
average_price_per_brand = pho1.groupby('Brand')['Discount price'].mean().sort_values(ascending=False)
print("\nAverage Prices by Brand:")
print(average_price_per_brand)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_per_brand.index, y=average_price_per_brand.values, palette='magma')
plt.title('Average Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(pho1['Actual price'].dropna(), bins=30, alpha=0.5, label='Actual Price')
plt.hist(pho1['Discount price'].dropna(), bins=30, alpha=0.5, label='Discount Price')
plt.title('Distribution of Actual Price and Discount Price')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(pho1['Stars'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating (Stars)')
plt.ylabel('Number of Products')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(pho1['Discount price'], pho1['Stars'], alpha=0.6, edgecolor='w', linewidth=0.5)
plt.title('Price vs. Rating')
plt.xlabel('Discount Price (INR)')
plt.ylabel('Rating (Stars)')
plt.grid(True)
plt.show()
```
