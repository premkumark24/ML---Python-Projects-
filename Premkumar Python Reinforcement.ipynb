```python
hous['sqft_living'].fillna(hous['sqft_living'].mean())
hous['sqft_lot'].fillna(hous['sqft_lot'].mean())
hous['yr_built'].fillna(hous['yr_built'].mean())
```

```python
hous['city'].fillna(hous['city'].mode()[0])
```

```python
'''4. Drop irrelevant columns such as Street, Statezip, and Country. Explain why these may
not directly affect housing price predicƟons.'''
hous = hous.drop(['street','statezip','country'], axis = 1)
hous
```

```python
#5. Create a new column House_Age = 2025 - Yr_built.
hous['House_Age'] = 2025 - hous['yr_built']
hous
```

```python
#6. Create another new column Price_per_sqŌ = Price / SqŌ_living.
hous['price_per_sqŌ'] = hous['price']/ hous['sqft_living']
hous
```

```python
hous['yr_built'].fillna(hous['yr_built'].mean())
hous['price_per_sqŌ'].fillna(hous['price_per_sqŌ'].mean())
```
